```python
# creating a function to detect outliers in any variable based on the IQR
def find_outliers(data):
  q1 = data.quantile(0.25)
  q3 = data.quantile(0.75)
  IQR = q3-q1
  outliers = data[((data<(q1-1.5*IQR)) | (data>(q3+1.5*IQR)))]
  return outliers
```
